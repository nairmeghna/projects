**About Dataset**

This data contains songs from 1950s. For each song, a set of attributes have been reported in order to perform some data analysis. In particular, the attributes are:

top genre: genre of the song

*   **top genre:** genre of the song
*   **year:** year of the song (due to re-releases, the year might not correspond to the release year of the original song)
*   **bpm(beats per minute):** beats per minute
*   **nrgy(energy):** energy of a song, the higher the value the more energetic the song is
*   **dnce(danceability):** the higher the value, the easier it is to dance to this song.
*   **dB(loudness):** the higher the value, the louder the song.
*   **live(liveness):** the higher the value, the more likely the song is a live recording.
*   **val(valence):** the higher the value, the more positive mood for the song.
*   **dur(duration):** the duration of the song.
*   **acous(acousticness):** the higher the value the more acoustic the song is.
*   **spch(speechiness):** the higher the value the more spoken word the song contains.
*  **pop(popularity):** the higher the value the more popular the song is.

```python
# importing necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# loading data
music = pd.read_csv("/1950.csv")
music
```

```python
# viewing data in detail
print(music.info())
print(music.describe())
```

```python
# setting the column 'nNumber' as the index
music.set_index('Number', inplace = True)

# dropping the 'year' column
music.drop(['year'], axis = 1, inplace = True)

music.head(5)
```

```python
# checking for null values if any
music.isna().sum()
```

```python
# removing null values
music.dropna()
```

**Analysis of The Genre**

```python
# finding the percent of each genre
genre = music['top genre'].value_counts()/len(music)
sizes = genre.values.tolist()
labels = genre.index.values.tolist()

# making a pie chart for genre
pfig = plt.figure(figsize = (10,10))
plt.pie(sizes, labels = labels, autopct='%1.1f%%')
```

From this pie chart it can be observed that more than 50% of the most popular songs from the dataset are from the genre 'adult standards'.

**Analysing Numeric Columns**

```python
# making a boxplot by excluding the non numeric columns
sns.boxplot(data = music.drop(['title', 'artist', 'top genre'], axis = 1))
plt.xlabel('Feature')
plt.ylabel('Value')
plt.show()
```

The boxplot highlights the presence of some outliers which can be dealt with either by removing the top 1% values or by removing the bottom 1%.

**Analysing the Top 3 Popular Genres**

```python
# top 3 popular genres
top = ['adult standards', 'brill building pop', 'deep adult standards']
music_ = music.loc[music['top genre'].isin(top)]
sns.pairplot(data = music_.drop(['title', 'artist'], axis = 1), hue = 'top genre')
```

It can be observed that most variables are linearly related, even if the datapoints are sparse.

**Correlation Matrix**

```python
# making a correlation matrix
fig, ax = plt.subplots(figsize = (15,10))
sns.heatmap(music.corr(), annot = True, cmap = 'YlGnBu', vmin = -1, vmax = 1, center = 0, ax = ax)
plt.title('Linear Correlation Matrix')
```

The correlation matrix confirms that most of the variables are positively related, i.e. they show a similar behaviour with only exceptions being 'dur' and 'acous'.
